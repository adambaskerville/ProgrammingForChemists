## <font color = '545AA&'> Basic Plotting </font>

Let's start with making a plot of sine and cosine. We will generate a series of number using the range() function and then calculate the sine and cosine of these values. This usually requires a `for` loop, but we can shorten our code using **list comprehension**, which is like putting a `for` loop in a single line.

```python
x = [num / 2 for num in range(1, 50)]
y = [math.sin(num)/num for num in x]

plt.plot(x, y, 'bo')
```

```python
plt.plot(x, y, 'b-')
```

This line plot is the most basic type of plot. If you look closely, you will noticed the verticies where the data points are located. However, it would be nice to be able to see the data points more clearly with **markers**.

We can change the style of the plot itself including markers, line types, and colors.

```python
plt.plot(x, y, 'ro') # ro means "red dot"
```

```python
plt.plot(x, y, 'ro-') # ro- means "red dot with a line"
```

### Saving Figures

It is likely that you will want to save your plot to a file for papers and presentations. You can do this with the **`plt.savefig()`** function.

    plt.savefig(fname, dpi = , format = )

The `fname` is the file name and is required. All other arguments are optional. `dpi` is the resolution in dpi (dots per inch). The format is the file format and should be given as a string (e.g., 'pdf', 'png', 'svg', etc...).

```python
plt.plot(x, y, marker = 'o', color = 'r', linestyle = '--')
plt.xlabel('x')
plt.ylabel('sin(x)/x')
plt.title('Sin(x)/x Plot')
plt.savefig('plot.png', dpi = 300, format = 'png')
```

Take a look in the directory where the Jupyter notebook is located to find your plot. If you are not running Python in a Jupyter notebook, the file will be saved in the same directory as your Python script or in the current active directory.

There are a number of other plotting types availale in matplotlib including **scatter plots, bar plots,** and **histograms**.

### Scatter Plots

Scatter plots are simply plots with markers but no lines connecting them. This can be done using `plt.plot()` with a marker but no line, so it is somewhat redundant.

The arguments after the `x` and `y` are a litte different than in `plt.plot()`. I personally use `plt.plot()` to make both line and scatter plots, but it's good to know that `plt.scatter()` is out there.

```python
plt.scatter(x, y)
```

### Bar Plots

Bar plots represent the same data as scatter plts except that there is a bar in the place of a dot or marker. The width of the bar has no meaning in a bar plot. Only the *left edge* on the x-axis and the height of the bar are important.

```python
domain = range(10)
squares = [x**2 for x in domain]

plt.bar(domain, squares)
```

### Histogram Plots

Historgram plots can be confused with bar plots because they look similar, but they represent data very differently. Bar plots represent pairs of (x, y) values while histograms represent the *frequency* of values within various intervals (i.e., bins). Also, the width of the bars in a bar plot are partially an aethetic choice whereas in a histogram plot, the width of the bars represent the bin size.

    plt.hist(data, bins)
    
Data is the raw, uncounted data and bins can either be an integer value telling matplotlib how many equally-sized bins to sort the data into or a list of bin cutoff values as a list, tuple, or array.

## <font color = '545AA7'> Multiple Plots in a Single Figure </font>

It is not uncommon that you will want to include multiple plots in a single figure. This is easily done by either calling the `plt.plot()` function multiple times in a row for each plot or my including the data for each plot in a single `plot.plot()` function call.

### Multiple Function Calls

```python
x = range(-50,51)
#x = [val for val in x if val]  # get ride of zero (method known as list comprehension)

y1 = []  # x^2
y2 = []  # -x^2
y3 = []  # x^2 * sin(x)

for val in x:
    y1.append(val**2)
    
for val in x:
    y2.append(-1 * val**2)
    
for val in x:
    y3.append(val**2 * math.sin(val))
```

```python
plt.plot(x, y1, 'g-', linewidth = 2, label = '$x^2$')
plt.plot(x, y2, 'b-', linewidth = 2, label = '$-x^2$')
plt.plot(x, y3, 'r-', linewidth = 2, label = '$x^2 \sin(x)$')
plt.xlabel('x')
plt.legend(loc='best')
```

### Single Function Call

```python
plt.plot(x, y1, 'g-', x, y2, 'b-', x, y3, 'r-', linewidth = 2)  # only had to call linewidth once
```
